```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings("ignore")
```

```python
#Extracting the data
data = pd.read_excel("omedena.xlsx")
data.head()
```

```python
# Checking if their is any missing data
data.isna().sum()
```

```python
#Plotting the scatter graph of the height and the date
fig,ax = plt.subplots(figsize = (10,5))
ax.scatter(data["height"],data["date"]);
```

```python
#Plotting histogram of the height
fig,ax = plt.subplots(figsize = (10,5))
ax.hist(data["height"]);
```

```python
#Plottimg the bar graph of the height and the date
fig,ax = plt.subplots(figsize = (10,5))
ax.bar(data["height"],data["date"]);
```

```python
#Plotting histogram of the error

fig,ax = plt.subplots(figsize = (10,5))
ax.hist(data['errors ']);
```

```python
# Plotting the scatter plot of the height and errors
fig,ax = plt.subplots(figsize=(10,5))
ax.scatter(data["height"],data['errors ']);
```
